```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scripts import figure_utils as utils
from scripts import lif_utils
```

Import data

```python
data = pd.read_pickle('.\\gridsearch_dataframe')
```

Constants

```python
COLOURS=['#380001', '#DEDE80', '#4CBD4C', '#239C9B', '#06087A', '#580059', '#FFC0C0']
NEURONS=[' 50', ' 100', ' 200', ' 500', ' 1000', ' 2000', ' 5000']
NEURON_PROP=[' 0.01', ' 0.1', ' 0.25', ' 0.5', ' 0.75', ' 0.9', ' 0.99']
```

### Box Plot

```python
my_pal = dict(zip(data['n_neurons '].unique(), COLOURS))

sns.set(rc={"figure.figsize":(6,2.5)}, style='ticks', font_scale=1)
flierprops = dict(markersize=1)
sns.boxplot(x='sparsity ', y="mean_rew", hue='n_neurons ', data=data, 
                 order=NEURON_PROP, hue_order=NEURONS, palette=my_pal,
           linewidth=0.5, flierprops=flierprops)

## Aesthetics
label_size = 10
tick_size = 9

plt.xlabel('Proportion of Neurons Active', fontsize = label_size)
plt.xticks(np.arange(7), NEURON_PROP, fontsize=tick_size)

plt.ylabel('Mean Reward', fontsize=label_size)
plt.yticks(fontsize=tick_size)

plt.legend(title='N Neurons', fontsize=tick_size, title_fontsize=label_size, loc='upper right',
          frameon=True)
plt.xlim(-0.5, 8)
plt.savefig('boxplot_sparsity.pdf', bbox_inches='tight', dpi=200)

plt.show()
```
